```python
from ultralytics import YOLO
import os
from google.colab import drive
import shutil

# Define la ruta de configuraci√≥n y carga el modelo
model = YOLO("yolov8s.yaml")  # Crea un modelo nuevo desde cero

# Configura los par√°metros de entrenamiento con early stopping
results = model.train(
    data=os.path.join(ROOT_DIR, "google_colab_config.yaml"),
    epochs=300,  # M√°ximo n√∫mero de epochs
    patience=20  # Det√©n si no mejora en 10 epochs consecutivos
)

# Copia los resultados al Google Drive
drive.mount('/content/drive')
shutil.copytree('/content/runs', '/content/drive/My Drive/Colab Notebooks/run25_11')
```
